## Задание 1 
### Даны 2 переменных, в которых хранятся строки произвольной длины: phrase_1 и phrase_2.
### Напишите код, который проверяет какая из этих строк длиннее.

```python
phrase_1 = 'Съешь же ещё этих мягких французских булок, да выпей чаю.'
phrase_2 = 'Съешь ещё этих мягких французских булок, да выпей же чаю.'
if len(phrase_1) > len(phrase_2):
    print ('Фраза 1 длиннее фразы 2')
elif len(phrase_1) < len(phrase_2):
    print('Фраза 1 короче фразы 2')
else:
    print('Фразы одинаковой длины')
```

width = 10

```python
# Вариант 1
x = input()
a = int(x[0])
b = int(x[1])
c = int(x[2])
d = int(x[3])
e = int(x[4])
f = int(x[5])
x1 = a + b + c
x2 = d + e + f
if x1 == x2:
    print('Билет счастливый')
else:
    print('Билет несчастливый')
```

## Задание 6 (необязательное)
### Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

если пользователь выбрал круг, запрашиваем его радиус,
если треугольник – длины трех его сторон;
если прямоугольник – длины двух его сторон.

```python
figure = input('Введите тип фигуры: ')
if figure == 'круг':
    rad = int(input('Введите радиус круга: '))
    print('Площадь этого круга: ', round(2 * rad * 3.1415, 2))

elif figure == 'треугольник':
    a = int(input('Введите длину стороны a: '))
    b = int(input('Введите длину стороны b: '))
    c = int(input('Введите длину стороны c: '))
    p = (a+b+c)/2
    S = (p * (p-a) * (p-b) * (p-c))**0.5
    if a+b<c or a+c<b or b+c<a:
        print('Такого треугольника не существует')
    else:
        print('Площадь этого треугольника: ', round(S, 2))
        
elif figure == 'прямоугольник':
    long = int(input('Введите длину стороны 1 '))
    short = int(input('Введите длину стороны 2 '))
    print('Площадь этого треугольника: ', long * short)
    
else:
    print('Я не знаю такую фигуру')
```
